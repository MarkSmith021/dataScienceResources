# HOW TO IMPORT MATPLOTLIB

##### "plt" is used as an alias here

```python
import matplotlib.pyplot as plt
```

# HOW TO PLOT A LINEAR GRAPH

##### "x" and "y" are array of points on the graph which are passed as arguments to plot() function to create a linear graph

```python
x = [1,2,3,4,5,6,7]
y = [2,4,6,8,10,12,14]
```

```python
plt.plot(x,y)
```

# REMOVE THE MEMORY ADDRESS LINE AND CHANGE THE COLOR TO RED

##### show() function removes the memory address from display

```python
plt.plot(x,y)
plt.show()
```

##### An additional attribute is passed to change the color. Here 'r' means red color. Writing "color" is not essential. Just 'r' as an argument would work too

```python
plt.plot(x,y, color = 'r')
plt.show()
```

```python
x = [1,2,3,1,5,6,7]
y = [2,4,6,8,10,2,14]
plt.plot(x,y,'g')
plt.show()
```

# SHOW THE MARKERS

#### A marker is a x and y coordinate on the graph

```python
x = [1,2,3,1,5,6,7]
y = [2,4,6,8,10,2,14]
plt.plot(x,y,'g', marker = "o")
plt.show()
```

# PLOT A BAR GRAPH AND CHANGE COLOR OF EACH BAR

##### bar() function is used to create bar diagrams

```python
x = [1,2,3,4]
y = [2,4,6,14]
clr = ['r','g','b','y']
plt.bar(x,y, color = clr)
plt.show()
```

# PLOT A HORIZONTAL BAR GRAPH

##### barh() function is used to create a horizontal bar diagram

```python
import numpy as np
x = np.array(['A','B','C','D'])
y = np.array([10,12,14,16])
plt.barh(x,y)
plt.show()
```

```python
plt.barh(x,y, color = 'red') #'r' / Color code would work as well
plt.show()
```

# CHANGE THE WIDTH OF BARS

```python
plt.bar(x,y, color = 'red', width = 0.1)
plt.show()
```

# HOW TO CREATE A SCATTERPLOT DIAGRAM

##### scatter() function is used to create a scatterplot diagram

```python
plt.scatter(x,y)
plt.show()
```

# HOW TO ADD A TITLE AND X AND Y LABELS TO THE DIAGRAM

##### title("titleName") function is used to display the title of a diagram
##### xlabel("xLabelName") function is used to display the x label of a diagram
##### ylabel("yLabelName") function is used to display the y label of a diagram

```python
clr = ["red","blue","magenta","pink"]
plt.scatter(x,y, color = clr)
plt.title("Scatter Plot Diagram")
plt.xlabel("Person")
plt.ylabel("Hours")
plt.show()
```

# HOW TO CHANGE THE SIZE OF THE DIAGRAM

##### Just like we can assign a value to "color" attribute to change the color of the diagram, we could assign a value to the "s" attribute to change the size of the diagram

```python
clr = ["red","blue","magenta","pink"]
plt.scatter(x,y, color = clr, s = 150)
plt.title("Scatter Plot Diagram")
plt.xlabel("Person")
plt.ylabel("Hours")
plt.show()
```

# USING PYTHON'S IMAGE LIBRARY [PILLOW] WITH MATPLOTLIB

##### Importing Image module from Pillow which would let us perform various operations on the desired image

##### Using Matplotlib to map the image to grayscale and converting it to ...

##### The imshow() module of pyplot is provided with two arguments: the image and the cmap effect we want

##### Just like graphs, pictures too can be assigned a title, x and y labels using pyplot's modules

##### pyplot's grid module could also be used to create a grid on the image

##### The modified image is then shown using pyplot's show() module

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

imagePath = r"imageOne.jpg"

image = Image.open(imagePath).convert('L') # RGB could be used too

plt.title("Image")

plt.grid()

plt.imshow(image, cmap = "gray") # coolmap/rainbow/etc could be used to

plt.show()
```

# HOW TO DRAW PIE CHARTS USING MATPLOTLIB

##### The pyplot's pie() module takes in a single argument to plot the pie chart. However to assign labels, we could assign "labels" attribute in the pie module as y

```python
import matplotlib.pyplot as plt

x = [36, 72, 100, 15, 20]
y = ["A","B","C","D","E"]

plt.pie(x)
plt.show()

plt.pie(x, labels = y)
plt.show()
```

##### To change the color of sectors, you could assign an array of colors to the "color" attribute in the pyplot's pie() module

```python
import matplotlib.pyplot as plt

x = [36, 72, 100, 15, 20]
y = ["A","B","C","D","E"]
clr = ["red","magenta","cyan","yellow","blue"]

plt.pie(x, labels = y, colors = clr)
plt.show()
```

##### We could create a key to identify what a sector reflects as well by using pyplot's legend() module

```python
import matplotlib.pyplot as plt

x = [36, 72, 100, 15, 20]
y = ["A","B","C","D","E"]
clr = ["red","magenta","cyan","yellow","blue"]

plt.pie(x, labels = y, colors = clr)
plt.legend()
plt.show()
```

##### To save the matplotlib's diagrams, use pyplot's savefig(filePath) method

```python
import matplotlib.pyplot as plt

x = [36, 72, 100, 15, 20]
y = ["A","B","C","D","E"]
clr = ["red","magenta","cyan","yellow","blue"]

plt.pie(x, labels = y, colors = clr)
plt.legend()
plt.savefig("image.jpg")
plt.show()
```
